```python
import pandas as pd#引入pandas套件
trainDf = pd.read_csv("train.csv")#讀取訓練資料
trainDf.head(10)#先印出前十筆資料看一下有哪些feature
```

```python
x_train_len=len(trainDf)#計算訓練資料的筆數
y_train = trainDf.iloc[:,-1]#拿出訓練資料的target
testDf = pd.read_csv("test.csv")#讀取測試資料
all_data = pd.concat([trainDf.iloc[:,1:-1],testDf.iloc[:,1:]])#合併訓練及測試資料的feature
all_data = pd.get_dummies(all_data)#對類別資料做one hot encoding
for i in range(len(all_data.iloc[0])):
    all_data.iloc[:,i] = all_data.iloc[:,i].fillna(all_data.iloc[:,i].median())#將空值補上那個欄位的中位數
all_data.head(10)#印出前十筆資料
```

```python
from sklearn.preprocessing import StandardScaler#引入標準化用的函式
standardScaler = StandardScaler()#生成標準化用的物件
standardScaler.fit(all_data)#訓練標準化用的模型
all_data = standardScaler.transform(all_data)#用訓練好的標準化模型轉換原本的資料
all_data[:5]#印出前五筆資料
```

```python
x_train = all_data[:x_train_len,:]#從合併的資料拿出訓練資料
x_test = all_data[x_train_len:,:]#從合併的資料拿出測試資料
```

```python
from xgboost import XGBRegressor#引入xgboost用的套件
from sklearn.model_selection import GridSearchCV#引入GridSearchCV用的套件
xgbRegressor =  XGBRegressor()
parameters = {'n_estimators':[200,300,400], 'max_depth':[3,4,5], 'reg_lambda':[0.1,0.2,0.3],'subsample':[0.7,0.8],'colsample_bytree':[0.7,0.8],'colsample_bylevel':[0.5,0.6,0.7,0.8]}#設定我要找的參數及範圍
gridSearchCV = GridSearchCV(xgbRegressor, parameters,cv=5)#將要訓練的模型及要找的參數和K-FOLD的數量給GridSearchCV
gridSearchCV.fit(x_train,y_train)#讓GridSearchCV用訓練資料找最佳參數
print(gridSearchCV.best_params_)#印出最佳參數確認一下
```

```python
xgbRegressor =  XGBRegressor(**(gridSearchCV.best_params_))#將最佳參數傳進xgbRegressor
xgbRegressor.fit(x_train,y_train)#訓練xgbRegressor
```

```python
testSalePrice = xgbRegressor.predict(x_test)#用訓練好之模型預測測試資料的target
testId = testDf['Id']
submission = pd.DataFrame({'Id':testId,'SalePrice':testSalePrice})#用DataFrame儲存結果以便輸出
submission.to_csv("result.csv",index = False)#將DataFrame內容輸出至csv
```
